```python
np.select() – Multiple Conditions at Once
Use when you have more than one condition to check.


import numpy as np

arr = np.array([10, 20, 30, 40])

conditions = [
    arr < 15,
    arr < 35
]

choices = ['Low', 'Medium']

np.select(conditions, choices, default='High')
➤ Output:
css

['Low', 'Medium', 'Medium', 'High']
2️⃣ df.apply() – Apply a function row-wise or column-wise

import pandas as pd

df = pd.DataFrame({'x': [1, 2, 3]})

df['squared'] = df['x'].apply(lambda val: val**2)
🧠 Think of it as: "Do something to every value".

✅ Works on entire rows too:


df.apply(lambda row: row['x'] * 10, axis=1)
3️⃣ df.map() – Apply simple functions to Series (1D)


df['label'] = df['x'].map({1: 'one', 2: 'two', 3: 'three'})
🧠 Use map() when you're replacing or transforming values in one column.

4️⃣ df.mask() – Opposite of where
Use when you want to replace values that meet a condition:



df['x'] = df['x'].mask(df['x'] > 2, 999)
📌 This means:

If value > 2 → replace with 999
Else → keep original value

5️⃣ df.clip() – Limit values within a range


df['x'].clip(lower=1, upper=2)
✂️ Any value < 1 becomes 1
✂️ Any value > 2 becomes 2

```
